```python
import pandas as pd
```

```python
column_names = [
    'movie_id',
    'movie_cluster',
    'movie_name',
    'release_date',
    'revenue',
    'runtime',
    'language',
    'country',
    'genres'
]

# Load the TSV file into a DataFrame with new column names
df_csu = pd.read_csv('movie.metadata.tsv', delimiter='\t', header=None, names=column_names)

display(df_csu)
```

```python
df_runtime= pd.read_csv('movies.csv')
df_wiki.drop_duplicates(['BookTitle','BookYear','Author','FilmTitle','FilmYear', 'full name', 'tmdb_id','id_goodreads','imdbid'], inplace=True)
print(df_wiki.duplicated(subset=['imdbid', 'BookTitle', 'FilmYear']).sum())  

df_runtime.drop_duplicates(['imdbid', 'title'], inplace=True)
print(df_runtime.duplicated(subset=['imdbid', 'title']).sum())  
```

```python
book_summaries = pd.read_csv('booksummaries_cleaned.csv', sep = ';',names =['book_ID', 'book_Code', 'book_title', 'author','year','details','book_summary'])
#book_summaries.head()
```

```python
merged_book_summaries = pd.merge(merged_movies_books, book_summaries, left_on='BookTitle',right_on='book_title', how='inner')
#Common_titles.head()
```

```python
set_df1 = set(merged_movies_books['BookTitle'])
set_df2 = set(book_summaries['book_title'])

# Find values that are not in both DataFrames
unique_to_df1 = set_df1 - set_df2
unique_to_df2 = set_df2 - set_df1

result_df = pd.DataFrame(unique_to_df2, columns=['Unique Values'])
mask = ~result_df.apply(lambda x: x.str.match(r'^[a-zA-Z]')) #r'^[a-zA-Z0-9]')

# Use the mask to filter the DataFrame
result_df2 = result_df[mask.any(axis=1)]

print("Values that don't start with any alphanumeric character:")
print(result_df2)
```
